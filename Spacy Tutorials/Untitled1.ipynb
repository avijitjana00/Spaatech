```python
from pdfminer.high_level import extract_text
import docx
import pathlib,os,csv
import pandas as pd
from pathlib import Path
```

```python
import docx2txt
header =['id','name','text']
count = 1
with open('resume.csv','w',encoding = 'UTF8',newline = '')as f:
    writer = csv.writer(f)
    writer.writerow(header)
    for path in pathlib.Path("resume").iterdir():
        if path.is_file():
            name,extension = os.path.splitext(path)
            base_path,f_name = os.path.split(path)
            if(extension == ".pdf"):
                text = extract_text(path)
                writer.writerow([count,f_name,text])
            elif(extension ==".docx"):
                text = docx2txt.process(path)
                writer.writerow([count,f_name,text])
            count += 1
    
```

```python
data = pd.read_csv('resume.csv')
data
```

```python
import numpy as np
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
from string import punctuation
_stopwords = set(stopwords.words('english') + list(punctuation))
l1 = []
for d in data['text']:
    d = sent_tokenize(d)
    s = [word_tokenize(i) for i in d]
    l = []
    for i in s:
        s1 =" "
        for j in i:
            if j not in _stopwords:
                s1 = s1 + j+" "
            l.append(s1)
    
    l1.append(l)
data['tokens']=l1
```

```python
import nltk,re
l2 = []
exp = ["experience","experienced","working","worked"]
for string in data['text']:
    x = []
    l = []
    string = string.lower()
    if "years" in string:
        l = re.findall(r"([^.]*?years[^.]*\.)",string)
        for i in l:
            if any(word in i for word in exp):
                pos_tags = nltk.pos_tag(word_tokenize(i))
                x = [j[0] for j in pos_tags if (j[1] == 'CD' and len(j[0])<=3)]
    if(len(x)>1):
        c = 0
        for i in x:
            if i[-1] == '+' or i[-1]=='x':
                i = i[:-1]
            if(i.isdigit()):
                c +=int(i)
        x = [c]
    if(len(x) == 0):
        x = [0]
    l2.append(x[0])
```

```python
data['experience'] = l2
```
